```python
import matplotlib.pyplot as plt
import torch
```

```python
x = torch.rand(4, 4,3)
x
```

```python
plt.imshow(x)
plt.show()
```

```python
x.size()
```

```python
y=torch.empty(2,2,3)
print(y)#结果
for i in range(1,x.size()[0],2):#行,步长为2
    for j in range(1,x.size()[1],2):  #列
        print(str(i)+","+str(j))
        y[int((i-1)/2),int((j-1)/2),0]=max(x[i,j,0],x[i-1,j,0],x[i,j-1,0],x[i-1,j-1,0])
        y[int((i-1)/2),int((j-1)/2),1]=max(x[i,j,1],x[i-1,j,1],x[i,j-1,1],x[i-1,j-1,1])
        y[int((i-1)/2),int((j-1)/2),2]=max(x[i,j,2],x[i-1,j,2],x[i,j-1,2],x[i-1,j-1,2])
print(y)
```

```python
plt.imshow(y)
plt.show()
```

```python
image_path=".\pic\lenna.png"
pic=plt.imread(image_path)
pic.shape
```

```python
plt.imshow(pic)
plt.show()
```

```python
new=torch.empty(256,256,3)
for i in range(1,pic.shape[0],2):
    for j in range(1,pic.shape[1],2):  
        new[int((i-1)/2),int((j-1)/2),0]=float(max(pic[i,j,0],pic[i-1,j,0],pic[i,j-1,0],pic[i-1,j-1,0]))
        new[int((i-1)/2),int((j-1)/2),1]=float(max(pic[i,j,1],pic[i-1,j,1],pic[i,j-1,1],pic[i-1,j-1,1]))
        new[int((i-1)/2),int((j-1)/2),2]=float(max(pic[i,j,2],pic[i-1,j,2],pic[i,j-1,2],pic[i-1,j-1,2]))
print(new)#max后是numpy.float类型
```

```python
pic1=new
plt.imshow(new)
plt.show()#max pooling
```

```python
new2=torch.empty(256,256,3)
for i in range(1,pic.shape[0],2):
    for j in range(1,pic.shape[1],2):  
        new2[int((i-1)/2),int((j-1)/2),0]=float(sum([pic[i,j,0],pic[i-1,j,0],pic[i,j-1,0],pic[i-1,j-1,0]]))/4
        new2[int((i-1)/2),int((j-1)/2),1]=float(sum([pic[i,j,1],pic[i-1,j,1],pic[i,j-1,1],pic[i-1,j-1,1]]))/4
        new2[int((i-1)/2),int((j-1)/2),2]=float(sum([pic[i,j,2],pic[i-1,j,2],pic[i,j-1,2],pic[i-1,j-1,2]]))/4
print(new2)
```

```python
pic2=new2
plt.imshow(new2)
plt.show()#mean pooling
```

```python
plt.subplot(1,2,1)
plt.imshow(pic1)
plt.title('max')
plt.subplot(1,2,2)
plt.imshow(pic2)
plt.title('mean')
plt.show()#mean更好
```
