```python
Filter=torch.tensor([[0.5, 1., 0.5],
        [0.5, 1., 0.5],
        [0.5, 1., 0.5]])

res=torch.zeros(3,3)
for i in range(1,x.size()[0]-1):
    for j in range(1,x.size()[1]-1):#除去外面一圈  
        print(str(i)+","+str(j))
        sum=0
        for m in range(1,3):
            for n in range(1,3):
                sum+=x[i+m-2,j+m-2]*Filter[m-1,n-1]
        sum/=9
        print(sum)
        res[i-1,j-1]=sum
print(res)
```

```python
plt.imshow(res)
plt.show()
```

```python
Filter=torch.tensor([[0.5, 1., 0.5],
        [0.5, 1., 0.5],
        [0.5, 1., 0.5]])
print(Filter)
#卷积核

for i in range(1,pic.shape[0]-1):
    for j in range(1,pic.shape[1]-1):  
        sum=0
        for m in range(1,3):
            for n in range(1,3):
                sum+=pic[i+m-2,j+m-2]*Filter[m-1,n-1].item()
        sum/=9
        pic[i,j]=sum
print(pic)
```
