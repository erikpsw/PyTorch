```python
import numpy as np
import matplotlib.pyplot as plt
import math

x=np.linspace(0,9,500)
l=9
def f(x):
    return x + 10 *np.sin(3*x) + np.cos(x)
plt.plot(x,f(x))
```

```python
N=1
x1=5
K=70
y=[]
for i in range(500):
    x2=x1+l*math.pow(math.e,-N/K)*(np.random.rand()-0.5)
    while(x2<0 or x2>9):
        x2=x1+l*math.pow(math.e,-N/K)*(np.random.rand()-0.5)
    x1=x2
    N=N+1
    y.append(x2)
plt.plot(y)
```

```python
#完全贪心算法
N=1
y=[]
for i in range(100):
    x2=x1+l*math.pow(math.e,-N/K)*(np.random.rand()-0.5)#左右随机游走
    while(x2<0 or x2>9):
        x2=x1+l*math.pow(math.e,-N/K)*(np.random.rand()-0.5)
    if(f(x2)<f(x1)):
        x1=x2
    N=N+1
    y.append(x1)
y[-1]
```

```python
plt.plot(x,f(x))
plt.scatter(y[-1],f(y[-1]))
```

```python
#退火
N=1
accept=0
y=[]
while(accept<10):
    x2=x1+l*math.pow(math.e,-N/K)*(np.random.rand()-0.5)
    while(x2<0 or x2>9):
        x2=x1+l*math.pow(math.e,-N/K)*(np.random.rand()-0.5)
    if(f(x2)<f(x1)):
        x1=x2
        accept+=1
    else:
        if(np.random.rand()<0.3):
            accept+=1
        else:
            accept=0
    N=N+1
    y.append(x1)
y[-1]
```
